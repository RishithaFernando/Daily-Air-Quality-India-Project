#### Handle missing values

```python
# Set 'Date' as index for time-based interpolation
df_cleaning.set_index('Date', inplace=True)

# Step 1: Interpolate and group-median for time-varying pollutants
time_sensitive_cols = ['PM2.5', 'NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3', 'AQI']
for col in time_sensitive_cols:
    if col in df_cleaning.columns:
        df_cleaning[col] = df_cleaning[col].interpolate(method='time')
        df_cleaning[col] = df_cleaning.groupby('City')[col].transform(
            lambda x: x.fillna(x.median() if not x.dropna().empty else np.nan)
        )

# Step 2: Group-median only for PM10 and NH3 (can be spiky or non-time-dependent)
for col in ['PM10', 'NH3']:
    if col in df_cleaning.columns:
        df_cleaning[col] = df_cleaning.groupby('City')[col].transform(
            lambda x: x.fillna(x.median() if not x.dropna().empty else np.nan)
        )

# Step 3: Group-median for sparse pollutants
for col in ['Benzene', 'Toluene', 'Xylene']:
    if col in df_cleaning.columns:
        df_cleaning[col] = df_cleaning.groupby('City')[col].transform(
            lambda x: x.fillna(x.median() if not x.dropna().empty else np.nan)
        )

# Step 4: Fill AQI_Bucket (categorical) using mode per city
if 'AQI_Bucket' in df_cleaning.columns:
    df_cleaning['AQI_Bucket'] = df_cleaning.groupby('City')['AQI_Bucket'].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown')
    )

# Step 5: Final fallback — global median or default value for any remaining NaNs
for col in df_cleaning.columns:
    if df_cleaning[col].isnull().sum() > 0 and df_cleaning[col].dtype != 'object':
        df_cleaning[col] = df_cleaning[col].fillna(df_cleaning[col].median())
    elif df_cleaning[col].isnull().sum() > 0 and df_cleaning[col].dtype == 'object':
        df_cleaning[col] = df_cleaning[col].fillna('Unknown')

# Reset index to return 'Date' as column
df_cleaning.reset_index(inplace=True)
```

### IQR Handling

####  IQR Handling for PM2.5
#### PM2.5 is a fine particulate pollutant that can spike due to firecrackers, industrial emissions, etc.
#### These spikes can skew the dataset, so we apply IQR to remove statistical outliers.

```python
Q1_pm25 = df_cleaning['PM2.5'].quantile(0.25)
Q3_pm25 = df_cleaning['PM2.5'].quantile(0.75)
IQR_pm25 = Q3_pm25 - Q1_pm25
lower_pm25 = Q1_pm25 - 1.5 * IQR_pm25
upper_pm25 = Q3_pm25 + 1.5 * IQR_pm25

outliers_pm25 = df_cleaning[(df_cleaning['PM2.5'] < lower_pm25) | (df_cleaning['PM2.5'] > upper_pm25)]
print(f" PM2.5 Outliers Removed: {len(outliers_pm25)}")
df_cleaning = df_cleaning[(df_cleaning['PM2.5'] >= lower_pm25) & (df_cleaning['PM2.5'] <= upper_pm25)]
```

#### IQR Handling for PM10
#### PM10 levels vary widely due to road dust, construction, and events like Diwali.
#### High variability can cause extreme values, so IQR helps maintain consistency.

```python
Q1_pm10 = df_cleaning['PM10'].quantile(0.25)
Q3_pm10 = df_cleaning['PM10'].quantile(0.75)
IQR_pm10 = Q3_pm10 - Q1_pm10
lower_pm10 = Q1_pm10 - 1.5 * IQR_pm10
upper_pm10 = Q3_pm10 + 1.5 * IQR_pm10

outliers_pm10 = df_cleaning[(df_cleaning['PM10'] < lower_pm10) | (df_cleaning['PM10'] > upper_pm10)]
print(f" PM10 Outliers Removed: {len(outliers_pm10)}")
df_cleaning = df_cleaning[(df_cleaning['PM10'] >= lower_pm10) & (df_cleaning['PM10'] <= upper_pm10)]
```

#### IQR Handling for AQI
#### AQI is a derived metric and can be extremely skewed if the input pollutants are out of bounds.
#### Removing AQI outliers ensures better modeling and visualization results.

```python
Q1_aqi = df_cleaning['AQI'].quantile(0.25)
Q3_aqi = df_cleaning['AQI'].quantile(0.75)
IQR_aqi = Q3_aqi - Q1_aqi
lower_aqi = Q1_aqi - 1.5 * IQR_aqi
upper_aqi = Q3_aqi + 1.5 * IQR_aqi

outliers_aqi = df_cleaning[(df_cleaning['AQI'] < lower_aqi) | (df_cleaning['AQI'] > upper_aqi)]
print(f" AQI Outliers Removed: {len(outliers_aqi)}")
df_cleaning = df_cleaning[(df_cleaning['AQI'] >= lower_aqi) & (df_cleaning['AQI'] <= upper_aqi)]
```

#### Outlier handling using IQR is not applied to certain variables for the following reasons:
#### 1. Carbon Monoxide (CO): CO levels are generally low and stable across most monitoring locations. 
####    Occasional high values may represent true pollution events rather than errors or anomalies, 
####    therefore removing outliers could discard valid data. Instead, missing values are handled using interpolation.
#### 2. Nitric Oxide (NO), Nitrogen Dioxide (NO2), and Sulfur Dioxide (SO2): These pollutants can exhibit genuine variability 
####    due to specific local factors such as industrial activity or traffic density. 
####    As a result, indiscriminate removal of outliers using IQR may eliminate meaningful extreme values. 
####    Visual inspection and domain knowledge should guide any outlier treatment in these cases.
#### 3. AQI_Bucket: This variable is categorical, representing air quality categories (e.g., ‘Good’, ‘Poor’) and not numerical. 
####    Therefore, IQR-based outlier detection is not applicable. This variable will be processed using appropriate categorical encoding methods.

```python


pollutant_columns = ['NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

def numerical_histogram_plotter(df, col):
    clean_col = df[col].dropna()

    if clean_col.nunique() < 2:
        print(f" Skipping '{col}' - not enough unique values.")
        return

   
    fig = plt.figure(figsize=(14, 6))

    
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.hist(clean_col, bins=50, color='skyblue', edgecolor='black')
    ax1.set_title(f'Distribution of {col}')
    ax1.set_xlabel(f'{col} (µg/m³)')
    ax1.set_ylabel('Frequency')
    ax1.grid(True)

    
    try:
        binned = pd.qcut(clean_col, q=3, labels=['Low', 'Moderate', 'High'])
        binned_counts = binned.value_counts().sort_index()

        ax2 = fig.add_subplot(1, 2, 2)
        binned_counts.plot(kind='pie', autopct='%1.2f%%', startangle=90,
                           colors=['green', 'orange', 'red'], ax=ax2)
        ax2.set_ylabel('')
        ax2.set_title(f'Pie Chart of {col} Levels')
    except ValueError:
        print(f" Pie chart skipped for '{col}' due to binning error.")

    plt.tight_layout()
    plt.show()


for col in pollutant_columns:
    numerical_histogram_plotter(df_cleaning, col)
```

##### Monthly Average AQI Trend – Delhi

To smooth out daily noise and highlight seasonal and long-term trends, this plot shows the average AQI per month for Delhi. It provides a clearer picture of overall pollution behavior across years.


```python
city = 'Delhi'
delhi_data = df_cleaning[df_cleaning['City'] == city].copy()


monthly_avg = delhi_data.groupby(['Year', 'Month'])['AQI'].mean().reset_index()


monthly_avg['Date'] = pd.to_datetime(monthly_avg['Year'].astype(str) + '-' + monthly_avg['Month'].astype(str) + '-01')

monthly_avg = monthly_avg.sort_values('Date')

plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_avg, x='Date', y='AQI', color='teal', linewidth=2.5, marker='o')


tick_positions = pd.date_range(start=monthly_avg['Date'].min(), end=monthly_avg['Date'].max(), freq='3MS')
plt.xticks(ticks=tick_positions, labels=[d.strftime('%b %Y') for d in tick_positions], rotation=45, fontsize=9)

plt.title(f'Average Monthly AQI Trend in {city}', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('AQI (Monthly Average)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=1.2, color='dimgray')
plt.tight_layout()
plt.show()
```
